```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('salary_data.csv')
df.drop(columns='Id',axis=1,inplace=True)
df.head()
df.isnull().sum()
```

```python
X_train=df['YearsExperience'].to_numpy(dtype=float)
Y_train=df['Salary'].to_numpy(dtype=float)
length=len(X_train)
```

```python
plt.scatter(df['YearsExperience'],df['Salary'])
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title("YearsExperience vs Salary")
plt.show()
```

```python
c=0
m=0
learning_rate=0.01
```

```python
X_train=(X_train-np.mean(X_train))/np.std(X_train)
Y_train=(Y_train-np.mean(Y_train))/np.std(Y_train)
```

```python
min_cost=10000
min_m=0
min_c=0
for i in range (100):
    Y_pred=(m*X_train)+c
    cost=np.mean(np.sum(Y_pred-Y_train)**2)/2
    error=Y_pred-Y_train
    m=m-learning_rate*np.mean(error*X_train)
    c=c-learning_rate*np.mean(error)
    if cost<min_cost:
        min_cost=cost
        min_m=m
        min_c=c
        print(f"Cost:{cost},{i}")
```

```python

plt.scatter(X_train, Y_train, color='r')
plt.plot(X_train,min_m*X_train+min_c,color='b')
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title("YearsExperience vs Salary")
plt.show()
```
